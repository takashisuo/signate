```python
import os
import pandas as pd
import plotly.graph_objects as go
```

```python
w_df = pd.read_csv(os.path.join(".", "data", "weather.csv"))
display(w_df.head())
display(w_df.shape)
```

```python
# weather の確認

grp_w = w_df.groupby("area")
for name, g_df in grp_w:
    #print(f"name:{name}, len:{len(g_df)}")
    if name == '青森':
        fig = go.Figure(data=go.Scatter(x=g_df["date"].apply(lambda x: str(x)), y=g_df["mean_temp"]))
        fig.update_xaxes(title="date")
        fig.update_yaxes(title="mean_temp")
        fig.update_layout(title="青森")
        fig.write_html("first_figure.html", auto_open=False)
```

```python
# train 確認
train_df = pd.read_csv(os.path.join("data", "train.csv"))
train_df.head()
```

```python
uniques = train_df["kind"].unique()
print(uniques)

grps = train_df.groupby('kind')

for name, g_df in grps:
    if name == "だいこん":
        fig = go.Figure(data=go.Scatter(x=g_df["date"].apply(lambda x: str(x)), y=g_df["mode_price"]))
        fig.update_xaxes(title="date")
        fig.update_yaxes(title="price")
        fig.update_layout(title=f"{name}: price")
        iplot(fig)
```
